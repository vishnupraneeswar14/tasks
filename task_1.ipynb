### Data Collection and Data Filtering  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#loading data
data= pd.read_csv('clustering_data.csv')
#dropping duplicates
data=data.drop_duplicates()
#droping nan
data=data.dropna()
#converting 'Latitude', 'Longitude' by seperating numbers from the data, then converting them to float
data['Latitude']=data['Latitude'].str.split().str[0]
data['Latitude']=data['Latitude'].str.split(pat='Â').str[0]
data['Latitude']=data['Latitude'].str.split(pat='-').str[0]
data['Latitude']=data['Latitude'].astype('float')
data['Longitude']=data['Longitude'].str.split().str[0]
data['Longitude']=data['Longitude'].str.split(pat='Â').str[0]
data['Longitude']=data['Longitude'].str.split(pat='-').str[0]
data['Longitude']=data['Longitude'].astype('float')
#removing nan values once again
data=data.dropna()
#extracting the data of homestate
a=data.loc[data['StateName']=="TAMIL NADU"]
#reseting the indices 
data=data.reset_index(drop=True)
a=a.reset_index(drop=True)
# Below we have hardcoded geographic boundaries(approx.) of Tamilnadu(obtained from web)
lat_min=8
lat_max=14
long_min=76
long_max=81
#converting latitude, longitude data into list(from pandas dataframe)
x=[]    # This contains Latitude of every point in Tamilnadu
y=[]    # This contains Longitude of every point in Tamilnadu
for i in range(a['Latitude'].count()):
    if lat_min<=a['Latitude'].iloc[i]<=lat_max and long_min<=a['Longitude'].iloc[i]<=long_max:
        x.append(a['Latitude'].iloc[i])
        y.append(a['Longitude'].iloc[i])        
#plotting data with latitude as y axis, longitude as x axis
plt.scatter(y,x,linewidth=0.1)
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

### K-Means

```python
avg_sil=[]
# giving colors to selected k random points    
col=np.array(['red', 'green', 'hotpink', 'blue', 'yellow', 'magenta', 'cyan', 'orange', 'purple', 'tan', 'peru', 'steelblue', 'navy', 'springgreen', 'olive', 'pink', 'gray', 'brown'])
zim=col.shape[0]
b_k=0
b_value=0
print("PFA Silhouette Scores for respective k values")

for k in range(2,zim+2): 
    if k==zim+1:
        k=b_k
        plt.scatter(y,x,marker='.',linewidth=3,color='black')
        plt.ylabel('Latitude')
        plt.xlabel('Longitude')
        print("\n")
        print("From Silhouette Scores, the best k value is",k,"\n")
    l=[]
    # to get k unique random points
    while(len(l)<k):
        xi=np.random.randint(0,len(x))
        c=0
        for i in range(len(l)):
            if l[i]==xi:
                c+=1
        if c==0:
            l.append(xi)        
    ## lists a,b contain y,x coordinates of k random points
    a=[]
    b=[]
    for i in range(k):
        a.append(x[l[i]])
        b.append(y[l[i]])
        
    ##colors contain k distinct colors
    colors=[]
    for i in range(k):
        colors.append(col[i])
    if k==b_k:
        plt.scatter(b,a,marker='^',linewidth=3,c=colors)   
        plt.ylabel('Latitude')
        plt.xlabel('Longitude')
    
    #round 1 of k means algorithm
    index=[]
    def classi(x,y,a,b):
        index=[]
        ## index contains index value of closest random point for each datapoint
        for i in range(len(x)):
            dist=[]
            for j in range(len(a)):
                p=(((x[i]-a[j])**2)+((y[i]-b[j])**2))**0.5
                dist.append(p)
            cal=dist[0]
            jin=0
            for j in range(len(a)):
                if dist[j]<cal:
                    cal=dist[j]
                    jin=j
            index.append(jin)
        return index     
    index=classi(x,y,a,b)     
    if k==b_k:
        plt.scatter(y,x, marker=".",c=col[index])  
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')        
    
    #round 2 of k means algorithm
    cent_x=[]
    cent_y=[]
    def centri_x(x,colors,index):
        cent_x=[]
        ##centroid calculation
        for i in range(len(colors)):
            x_d=0
            count=0
            for j in range(len(index)):        
                if i==index[j]:
                    x_d+=x[j]
                    count+=1
            cent_x.append(x_d/count)
        return cent_x
    def centri_y(y,colors,index):
        cent_y=[]
        ##centroid calculation
        for i in range(len(colors)):
            y_d=0
            count=0
            for j in range(len(index)):        
                if i==index[j]:
                    y_d+=y[j]
                    count+=1
            cent_y.append(y_d/count)
        return cent_y    
    cent_x=centri_x(x,colors,index)
    cent_y=centri_y(y,colors,index)
    d_l=[]
    for i in range(k):
        d_l.append(i)
    if k==b_k:
        print("Current Centroidal Locations:")
        print(cent_y,cent_x)
        plt.scatter(cent_y,cent_x, marker="x", linewidth=3, c=col[d_l])
        plt.ylabel('Latitude')
        plt.xlabel('Longitude')
        plt.show()
        
    
    #round 3 of k means algorithm
    index_1=[]
    def classi_1(x,y,cent_x,cent_y):
        index_1=[]
        for i in range(len(x)):
            dist_1=[]
            for j in range(len(cent_x)):
                p_1=(((x[i]-cent_x[j])**2)+((y[i]-cent_y[j])**2))**0.5
                dist_1.append(p_1)
            cal_1=dist_1[0]
            jin_1=0
            for j in range(len(cent_x)):
                if dist_1[j]<cal_1:
                    cal_1=dist_1[j]
                    jin_1=j
            index_1.append(jin_1) 
        return index_1
    index_1=classi_1(x,y,cent_x,cent_y) 
    if k==b_k:
        plt.scatter(y,x, marker=".",c=col[index_1])  
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
    cent_x=[]
    cent_y=[]
    cent_x=centri_x(x,colors,index_1)
    cent_y=centri_y(y,colors,index_1)
    if k==b_k:
        print("Current Centroidal Locations:")
        print(cent_y,cent_x)
        plt.scatter(cent_y,cent_x, marker="x", linewidth=3, c=col[d_l])    
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
        plt.show()
    
    while(index!=index_1):
        index=index_1
        index_1=[]
        index_1=classi_1(x,y,cent_x,cent_y) 
        if k==b_k:
            plt.scatter(y,x, marker=".",c=col[index_1])    
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
        cent_x=[]
        cent_y=[]
        cent_x=centri_x(x,colors,index_1)
        cent_y=centri_y(y,colors,index_1)
        if k==b_k:
            print("Current Centroidal Locations:")
            print(cent_y,cent_x)
            plt.scatter(cent_y,cent_x, marker="x", linewidth=3, c=col[d_l])    
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
            plt.show()       
        
    # Finding the best possible k value using silhouette analysisl
    g=[]   # a[i] list as per analysis
    h=[]   # b[i] list as per analysis
    max=[] # This contains max(a[i],b[i])
    sil=[] # This contains silhouette score of all points
    
    # a[i] calculation
    for l in range(len(x)):
        count=0
        d=0
        for m in range(len(x)):
            if index_1[l]==index_1[m]:
                d+=((((x[l]-x[m])**2)+((y[l]-y[m])**2))**0.5)
                count+=1
        d/=count
        g.append(d)
    # b[i] calculation
    for l in range(len(x)):
        dummy=[]
        for m in range(len(colors)):
            d=0
            count=0
            for n in range(len(x)):
                if m==index_1[n]:
                    d+=((((x[l]-x[n])**2)+((y[l]-y[n])**2))**0.5)
                    count+=1
            d/=count
            dummy.append(d)
        min_dummy=10000000000000000    
        for i in range(len(dummy)):
            if i!=index_1[l]:
                if dummy[i]<min_dummy:
                    min_dummy=dummy[i]
        h.append(min_dummy)             
            
    #max[i] calculation
    for i in range(len(x)):
        if g[i]>h[i]:
            max.append(g[i])
        else:
            max.append(h[i])
    #silhouette score
    for i in range(len(x)):
        gip=(h[i]-g[i])/max[i]
        sil.append(gip)
    #avg silhouette score calculation
    avg=0
    for i in range(len(x)):
        avg+=sil[i]
    avg/=len(x)    
    avg_sil.append(avg)
    print(avg,": k=",k)
    
    if k==2:
        b_k=k
        b_value=avg
    if avg>b_value:
        b_k=k
        b_value=avg
```
